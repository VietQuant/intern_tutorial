# Overview

Datacollector chứa các loại data sau:
- universal  : **(data/metadata/universal)**
    - top 30, 50, 100 cổ phiếu có khối lượng giao dịch lớn nhất hàng tháng
- intraday data  : **(data/OS or data/IS)**
    - OHLCV(adjusted) theo phút
- Daily data  : **(data/live_data/daily)**
    - OHLCV(adjusted) theo ngày
- FDMT data  : **(data/fdmt)**
    - báo cáo tài chính
- sts data  : **(data/VIETSTOCK_STS)**
    - thống kê theo ngày
- unadjusted data  : **(data/live_data/unadjusted)**
    - Giá chưa điều chỉnh

## Intraday data

```python
from datacollector.candle_crawlers.vietstock_db_intraday_crawler import VietStockDBIntradayCrawler
from rework_backtrader.utils.external_request import get_all_symbols
get_all_symbols?
```

```python
# Get list stock
stocks = get_all_symbols()
stocks = ['VN30F1M'] + stocks
stocks[:8]
```

```python
IS_abs_folder = os.path.join(folder, 'data/IS')
print("IS Data se duoc chua o : ", IS_abs_folder)
OS_abs_folder = os.path.join(folder, 'data/OS')
print("OS Data se duoc chua o : ", OS_abs_folder)
```

```python
# Lay data IS, dùng 4 thread, thoi gian tu 2018-01-01 toi 2019-12-31
VietStockDBIntradayCrawler(4, output_folder=IS_abs_folder).run(names=stocks, from_time='2018-01-01', to_time='2019-12-31')
# Lay data OS, dùng 3 thread, thoi gian tu 2020-01-01 toi hien tai
VietStockDBIntradayCrawler(3, output_folder=OS_abs_folder).run(names=stocks, from_time='2020-01-01')
```
